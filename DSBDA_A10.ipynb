Import necessary libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load the dataset

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv(url, names=columns)
```

Check the contents of the dataset

```python
print(iris.head())
```

```python
print("\nFeatures and their types:")
print(iris.dtypes)
```

```python
# Create histograms for each feature
plt.figure(figsize=(10, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.histplot(iris[feature], kde=True)
    plt.title(f'Histogram of {feature}')

plt.tight_layout()
plt.show()
```

The histograms show the distribution of each feature in the Iris dataset

```python
# Create boxplots for each feature
plt.figure(figsize=(10, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=iris[feature])
    plt.title(f'Boxplot of {feature}')

plt.tight_layout()
plt.show()
```

The boxplots show the distribution and spread of each feature in the Iris dataset.

```python
# Identify outliers
print("\nOutliers:")
for feature in iris.columns[:-1]:
    Q1 = iris[feature].quantile(0.25)
    Q3 = iris[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = iris[(iris[feature] < lower_bound) | (iris[feature] > upper_bound)]
    print(f"{feature}: {outliers.shape[0]} outliers")
```

Observations:
1. Features are numeric: sepal length, sepal width, petal length, petal width.
2. Sepal width has a relatively normal distribution, while sepal length, petal length, and petal width are right-skewed.
3. There are outliers in sepal length and sepal width, but none in petal length and petal width.
4. Sepal length has the widest range of values, while petal length has the smallest range.
