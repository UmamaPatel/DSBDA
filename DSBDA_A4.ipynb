Create a Linear Regression Model using Python/R to predict home prices using Boston Housing Dataset (https://www.kaggle.com/c/boston-housing). The Boston Housing dataset contains information about various houses in Boston through different parameters. There are 506 samples and 14 feature variables in this dataset. The objective is to predict the value of prices of the house using the given features.

Importing necessary libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

Load the dataset

```python
url = "https://raw.githubusercontent.com/dphi-official/Datasets/master/Boston_Housing/Training_set_boston.csv"
data = pd.read_csv(url)
```

Check the contents of the dataset

```python
# Histogram of the target variable ('medv')
plt.figure(figsize=(8, 6))
sns.histplot(data['MEDV'], bins=30, kde=True)
plt.title('Histogram of Median Home Value')
plt.xlabel('Median Home Value ($1000s)')
plt.ylabel('Frequency')
plt.show()
```

Model Selection

```python
model = LinearRegression()
```

Define the features (X) and target (y) variables.

```python

X = data.drop('MEDV', axis=1)
y = data['MEDV']
```

Split the data into training and testing sets using train_test_split().

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)
```

Fit a linear regression model on the training data

```python
model.fit(X_train, y_train)
```

Evaluate the model on the test data and calculate the R-squared score using model.score()

```python
score = model.score(X_test, y_test)
print(score)
```

Observation: The model achieved a certain R-squared score on the testing data, indicating how well it fits the test data.
