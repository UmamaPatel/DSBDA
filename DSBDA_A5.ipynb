```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

Model Selection

```python
model = LogisticRegression()
```

Define the features (X) and target (y) variables.

```python
X = data[['Age', 'EstimatedSalary', 'Gender']]
y = data['Purchased']
```

Split the data into training and testing sets using train_test_split().

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

Fit a linear regression model on the training data

```python
model.fit(X_train, y_train)
```

Evaluate the model on the test data and calculate the R-squared score using model.score()

```python
score = model.score(X_test, y_test)
print(score)
```

The model achieved an R-squared score on the testing data, indicating how well it fits the test data.
