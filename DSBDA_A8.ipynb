Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset

```python
data = sns.load_dataset('titanic')
```

Check the contents of the dataset

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

Data Cleaning

```python
data['age'] = data['age'].fillna(np.mean(data['age']))
data['deck'] = data['deck'].fillna(data['deck'].mode()[0])
data['embark_town'] = data['embark_town'].fillna(data['embark_town'].mode()[0])
data.isnull().sum()
```

Data Visualization

1. Create a categorical plot showing the count of passengers who survived or not based on their gender ('sex').

```python
sns.catplot(x ="sex", hue ="survived",
kind ="count", data = data)
```

From the graph, it's clear that around 20% of men and about 75% of women survived the Titanic disaster. This indicates that gender played a significant role in determining survival rates.


2. Create a line plot showing the relationship between 'age' and 'fare' (ticket price).

```python
sns.lineplot(x='age', y='fare', data=data)
plt.show()
```

By looking at the line plot, we can interpret how the ticket fare varies with age. For example, if the line slopes upward, it indicates that as age increases, the fare tends to increase as well. If the line is relatively flat, it suggests that there may not be a strong relationship between age and fare.

3. Create a histogram showing the distribution of ticket prices ('fare').

```python
sns.histplot(x='fare', data=data)
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

This histogram shows the distribution of ticket prices among passengers, giving an idea of the range and frequency of ticket prices.

4. Create a joint plot showing the relationship between 'age' and 'fare'.

```python
sns.jointplot(x='age', y='fare', data=data,)
plt.show()
```

Scatter Plot: Each point represents a passenger, with their age on the x-axis and ticket fare on the y-axis.

Histograms: The histogram on the top shows the distribution of ages among passengers, while the histogram on the right shows the distribution of ticket fares.

5. Create a rug plot showing the distribution of ages.**bold text**

```python
sns.rugplot(x='age', data=data)
plt.show()
```

This rug plot shows how ages are distributed in the Titanic dataset. Each tick mark represents a passenger's age, so more ticks in an age range mean more passengers of that age. Denser areas with more ticks indicate where ages are more concentrated, while sparser areas with fewer ticks show where ages are less common.

6. Create a distribution plot of 'fare'.

```python
sns.distplot(data['fare'])
plt.show()
```
